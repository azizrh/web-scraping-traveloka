```python
# querying data from html
name_rating = soup.findAll(True, {'class':['fde444d7ef _c445487e2','af1ddfc958 eba89149fb','_9c5f726ff _192b3a196 f1cbb919ef','_9c5f726ff bd528f9ea6','fde444d7ef _e885fdc12','_4abc4c3d5 _1e6021d2f _6e869d6e0']})
list = []
for nr in name_rating:
    result = nr.text.strip()
    #print(result)
    list.append(nr.text.strip())
    
    #querying name
name=[]
i=0
while i<len(list):
    name.append(list[i])
    #print(list[i])
    i=i+7
    
    #querying location
loc=[]
i=1    
while i<len(list):
    #print(list[i])
    loc.append(list[i])
    i=i+7
    
    #querying rating
rating=[]
i=3    
while i<len(list):
    #print(list[i])
    rating.append(list[i])
    i=i+7
    
    #querying quality
quality=[]
i=4   
while i<len(list):
    #print(list[i])
    quality.append(list[i])
    i=i+7
    
    #querying price
reviews=[]
i=5   
while i<len(list):
    #print(list[i])
    reviews.append(list[i])
    i=i+7
    
    #querying number of reviews
price=[]
i=6   
while i<len(list):
    #print(list[i])
    price.append(list[i])
    i=i+7

df = pd.DataFrame(({'name': name, 'location': loc,'rating':rating,'quality':quality,'number of reviews':reviews,'price':price}))
df
```
